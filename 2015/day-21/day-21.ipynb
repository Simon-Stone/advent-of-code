Hit Points: 109
Damage: 8
Armor: 2

```python
from itertools import product
```

```python
with open("input.txt") as f:
    store = f.read()

store = list(
    map(
        lambda x: x.split(
            "\n",
        ),
        store.split("\n\n"),
    )
)
```

```python
WEAPONS = {
    name: {"cost": int(cost), "damage": int(damage), "armor": int(armor)}
    for name, cost, damage, armor in map(str.split, store[0][1:])
}
```

```python
ARMOR = {
    name: {"cost": int(cost), "damage": int(damage), "armor": int(armor)}
    for name, cost, damage, armor in map(str.split, store[1][1:])
}
```

```python
RINGS = {
    " ".join(name): {"cost": int(cost), "damage": int(damage), "armor": int(armor)}
    for *name, cost, damage, armor in map(str.split, store[2][1:])
}
```

```python
def can_beat_boss(w, a, rl, rr):
    hp_boss = 109
    damage_boss = 8
    armor_boss = 2
    hp_player = 100

    damage_player = WEAPONS[w]["damage"]
    if rl is not None:
        damage_player += RINGS[rl]["damage"]
    if rr is not None:
        damage_player += RINGS[rr]["damage"]
    armor_player = 0
    if rl is not None:
        armor_player += RINGS[rl]["armor"]
    if rr is not None:
        armor_player += RINGS[rr]["armor"]
    if a is not None:
        armor_player += ARMOR[a]["armor"]

    n_turns = int(hp_boss / max(damage_player - armor_boss, 1))

    return hp_player > n_turns * max(damage_boss - armor_player, 1)
```

```python
def calc_cost(w, a, rl, rr):
    cost = WEAPONS[w]["cost"]
    if a is not None:
        cost += ARMOR[a]["cost"]
    if rl is not None:
        cost += RINGS[rl]["cost"]
    if rr is not None:
        cost += RINGS[rr]["cost"]
    return cost
```

```python
weapons = list(WEAPONS.keys())
armor = [None] + list(ARMOR.keys())
rings = [None] + list(RINGS.keys())
```

```python
winning_costs = []
for w, a, rl, rr in product(weapons, armor, rings, rings):
    if rl == rr:
        continue
    if can_beat_boss(w, a, rl, rr):
        winning_costs.append(calc_cost(w, a, rl, rr))
min(winning_costs)
```

```python
losing_costs = []
for w, a, rl, rr in product(weapons, armor, rings, rings):
    if rl == rr:
        continue
    if not can_beat_boss(w, a, rl, rr):
        losing_costs.append(calc_cost(w, a, rl, rr))
max(losing_costs)
```
